# **1.Perulangan dan Percabangan**

**Perulangan**

For Loop

```python
for item in [1, 2, 3]:
  print(item)
```

While Loop

```python
count = 1
while count <= 3:
  print(count)
  count += 1
```

**Percabangan**

If-Else

```python
number = 4
if number % 2 == 0:
  print("Genap")
else:
  print("Ganjil")
```

If-Elif-Else

```python
score = 85
if score >= 90:
  print("A")
elif score >= 80:
  print("B")
else:
  print("C")
```

# **2. Strategi Algoritma Brute Force**

Mencari Nilai Maksimum

```python
def max_value(arr):
  max_val = arr[0]
  for num in arr[1:]:
    if num > max_val:
      max_val = num
  return max_val
print(max_value([3, 5, 1, 9, 2]))
```

Memeriksa Bilangan Prima

```python
def is_prime(n):
  if n < 2:
    return False
  for i in range(2, int(n ** 0.5) + 1):
    if n % i == 0:
      return False
  return True
print(is_prime(17))
```

# **5. Latihan Pendukung**

Tulis algoritma brute force untuk menghitung jumlah bilangan genap dalam list.

```python
def hitung_genap(lst):
    jumlah = 0
    for angka in lst:
        if angka % 2 == 0:
            jumlah += 1
    return jumlah

data = [1, 2, 3, 4, 5, 6]
print("Jumlah bilangan genap:", hitung_genap(data))
```

Buat diagram alir untuk algoritma yang menentukan apakah sebuah bilangan positif, negatif, atau nol.

```python
angka_list = [5, -2.5, 0, 8, -9]

for angka in angka_list:
    if angka > 0:
        print(f"{angka} adalah bilangan positif")
    elif angka < 0:
        print(f"{angka} adalah bilangan negatif")
    else:
        print(f"{angka} adalah bilangan nol")
```
